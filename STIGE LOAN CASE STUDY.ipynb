```python
# woah! since these are lot of null values.
# we need something to seee it better.
# Let's run off some analysis
sns.boxplot(data=app,x='AMT_GOODS_PRICE')
```

## 5. 'NAME_TYPE_SUITE' analysis

```python
# checking for null values
app['NAME_TYPE_SUITE'].isna().sum()
```

```python
# since it is a categorical variable, and null % is almost 0.42% which is quite a low.
# it will be good to replace the value with something better such as frequent value i.e. mode.
# we could have also replaced it with some other value such as 'NF','UNKNOWN' or 'OTHER'
# but mode is technically great.
app.NAME_TYPE_SUITE = app.NAME_TYPE_SUITE.fillna(app.NAME_TYPE_SUITE.mode()[0])
```

```python
# checking for null values
app['NAME_TYPE_SUITE'].isna().sum()
```

```python
# Converting categorical values from application data to categorical Dtype


col = ['NAME_CONTRACT_TYPE','CODE_GENDER','NAME_TYPE_SUITE','NAME_INCOME_TYPE','NAME_EDUCATION_TYPE',
                       'NAME_FAMILY_STATUS','NAME_HOUSING_TYPE','OCCUPATION_TYPE','WEEKDAY_APPR_PROCESS_START',
                       'ORGANIZATION_TYPE','FLAG_OWN_CAR','FLAG_OWN_REALTY','LIVE_CITY_NOT_WORK_CITY',
                       'REG_CITY_NOT_LIVE_CITY','REG_CITY_NOT_WORK_CITY','REG_REGION_NOT_WORK_REGION',
                       'LIVE_REGION_NOT_WORK_REGION','REGION_RATING_CLIENT','WEEKDAY_APPR_PROCESS_START',
                       'REGION_RATING_CLIENT_W_CITY'
                      ]
for a in col:
    app[a] =pd.Categorical(app[a])

```

1. Females applying for loan tend to make more default than males, however the difference is slight.

4. According to total data, Applicants with no children likes to default more.

6. The main Credit Cap for defaulters are between 500000 and 700000.

10. Working people defaults more than other Income type.

12. People working as lobourers are likely to default.
13. People working for Business Entity 3 have more chances to default.
14. AMT_INCOME,AMT_CREDIT,AMT_ANNUITY and AMT_GOOD_PRICE, these all values are correlated. They share positive relation though. But AMT_CREDIT with AMT_ANNUITY and AMT_GOOD_PRICE shows the most positive uprising.
